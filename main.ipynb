```python
from shapely.geometry import Point
from IPython.display import display, clear_output, Markdown

from Test.utils.Schema.output_schema import Plan, Node

def flatten_nodes(node: Node):
    """
    递归函数，用于展平读取每个装备或人员的名称和坐标
    """
    if node.nodetype in ['EquipNode', 'PersonnelNode']:
        if isinstance(node.position, tuple):
            yield (node.name, node.position)
    if node.children is not None and node.children != 'null':
        for child in node.children:
            yield from flatten_nodes(child)

def extract_names_and_positions(plan: Plan):
    """
    提取 Plans 对象中每个装备或人员的名称和坐标
    """
    result = []
    for node in plan.formation:
        result.extend(flatten_nodes(node))
    return result



generation = event.get('final_response')

def my_colormap(value):
    if value == '蓝方':
        return "blue"
    elif value == '红方':
        return "red"
    else:
        return "green"

marker_kwds = {'radius':10, 'draggable':True, 'fill':True}

teams = []

d={'team':[], 'name':[], 'geometry':[]}
if generation.red!='null' and generation.red!=None:
    color_str = '红方'
    teams.append((color_str, generation.red))
if generation.blue!='null' and generation.blue!=None:
    color_str = '蓝方'
    teams.append((color_str, generation.blue))

for color, team in teams:
    clear_output()
    for name, pos in extract_names_and_positions(team):
        d['team'].append(color)
        d['name'].append(name)
        d['geometry'].append(Point(pos))
        # d['reason'].append(entity['steps'])
    gdf = gpd.GeoDataFrame(d, crs="EPSG:4326").sort_values(by='team', ignore_index=True, kind='stable')
    display(Markdown(f'## <p style="text-align: center;">**红蓝方部署**</p>'))
    gdf.geometry.set_precision(10)
    display(gdf)
    display(gdf.explore(column='team', cmap=[my_colormap(value) for value in gdf['team'].unique()], marker_type='circle_marker', marker_kwds = marker_kwds, tooltip = ['team', 'name']))
        # gdf.to_file(f'./plans/plan_{i}.geojson', driver='GeoJSON')
```
