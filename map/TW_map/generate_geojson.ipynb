# 下载地图

## 下载地图（online）

```python
import osmnx as ox
import geopandas as gpd

# place = "Manhattan"  # 地名
place = "桃園市"  # 地名
tags_building = {"building": True}
tags_highway = {"highway": True}
tags_shop = {"shop": True}
tags_amenity = {"amenity": True}

# boundary = ox.geocode_to_gdf(place)['geometry'][0]

gdf_highway = ox.features_from_place(place, tags_highway)
gdf_shop = ox.features_from_place(place, tags_shop)
gdf_amenity = ox.features_from_place(place, tags_amenity)
gdf_building = ox.features_from_place(place, tags_building)
```

## 加载地图（offline）

```python
import geopandas as gpd
gdf_highway = gpd.read_file('TW_yangmei_highway_zh.geojson')
gdf_shop = gpd.read_file('TW_yangmei_shop_zh.geojson')
gdf_amenity = gpd.read_file('TW_yangmei_amenity_zh.geojson')
gdf_building = gpd.read_file('TW_yangmei_building_zh.geojson')
```

# highway

```python
# gdf_highway = gdf_highway[gdf_highway.name.notna()][['name', 'highway', 'geometry']]
gdf_highway = gdf_highway[['name', 'highway', 'geometry']]
gdf_highway.rename(columns={'highway': 'tags'}, inplace=True)
gdf_highway.reset_index(drop=True, inplace=True)
gdf_highway
```

# shop

```python
gdf_shop
```

```python
# gdf_shop = gdf_shop[gdf_shop.name.notna()][['name', 'shop', 'geometry']]
gdf_shop = gdf_shop[['name', 'shop', 'geometry']]  
gdf_shop.rename(columns={'shop': 'tags'}, inplace=True)
gdf_shop.reset_index(drop=True, inplace=True)
gdf_shop.tags = gdf_shop.apply(lambda x: x['tags']+'-shop', axis=1)
gdf_shop
```

# amenity

```python
# gdf_amenity = gdf_amenity[gdf_amenity.name.notna()][['name', 'amenity', 'geometry']]
gdf_amenity = gdf_amenity[['name', 'amenity', 'geometry']]
gdf_amenity.rename(columns={'amenity': 'tags'}, inplace=True)
gdf_amenity.reset_index(drop=True, inplace=True)
gdf_amenity
```

# building

```python
# gdf_building = gdf_building[gdf_building.name.notna()][['name', 'building', 'geometry']]
gdf_building = gdf_building[['name', 'building', 'geometry']]
gdf_building.rename(columns={'building': 'tags'}, inplace=True)
gdf_building.reset_index(drop=True, inplace=True)
gdf_building.tags = gdf_building.apply(lambda x: 'common-building' if x['tags']=='yes' or x['tags']=='no' else x['tags'], axis=1)
gdf_building
```

# 裁剪地图

```python
from shapely import Polygon
import pandas as pd
map_data = pd.concat([gdf_highway, gdf_shop, gdf_amenity, gdf_building])
bbox = (121.14323, 24.903136, 121.15039, 24.907623)
rect = Polygon([(bbox[0], bbox[1]), (bbox[2], bbox[1]), (bbox[2], bbox[3]), (bbox[0], bbox[3]), (bbox[0], bbox[1])])
clip = map_data.clip(rect)
clip.explore()
```

## 保存裁剪文件

```python
clip.to_file('TW_yangmei_map_complete.geojson', driver='GeoJSON')
```
